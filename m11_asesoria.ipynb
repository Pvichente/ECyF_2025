## Algunos de los términos clave son:

Antes de sumergirnos en la tokenización, aclaremos algunos términos que se encuentran con frecuencia en NLP:

- **Corpus**: un cuerpo de texto, en singular. "Corpora" es la forma plural. Un corpus es un conjunto grande y estructurado de textos utilizados para realizar análisis lingüísticos y entrenar modelos de procesamiento del lenguaje natural (NLP). Puede consistir en textos escritos, transcripciones de discursos, o cualquier otro tipo de datos textuales. Un corpus puede ser general (contener textos de diversos temas y estilos) o especializado (con textos de un área específica, como medicina, derecho, o redes sociales).

- **Léxico**: en esencia, un diccionario que incluye palabras y sus significados.Un léxico se refiere al inventario de palabras o términos que aparecen en un corpus o conjunto de datos textuales, a menudo con información adicional como su significado, frecuencia o relaciones con otras palabras.

- **Token**: una pieza o "entidad" individual que resulta de descomponer el texto en función de reglas específicas.  Usualmente una palabra, pero también puede ser una sub-palabra, un carácter o una secuencia de caracteres, que es el resultado de un proceso de tokenización. Por ejemplo, una oración se puede tokenizar en palabras, convirtiendo cada palabra en un token. De manera similar, los párrafos se pueden tokenizar en oraciones, convirtiendo cada oración en un token.

```python
def grafica_longitud_documentos(
    df,
    columna="len_doctos",
    bins=30,
    titulo="Distribución de longitud de documentos"
):
    """
    Grafica la distribución de la longitud de documentos y su media.
    """
    # Calcular media
    media = df[columna].mean()

    # Crear figura
    plt.figure(figsize=(10, 6))

    # Histograma
    plt.hist(df[columna], bins=bins, color="skyblue", edgecolor="black", alpha=0.8)
    # Línea de la media
    plt.axvline(media, color="red", linestyle="--", linewidth=2,
                label=f"Media = {media:.2f}")
    # Detalles
    plt.xlabel("Número de caracteres")
    plt.ylabel("Frecuencia")
    plt.title(titulo)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

# Tokenización en NLP

La tokenización en el procesamiento del lenguaje natural (PNL) es una técnica fundamental que implica dividir una oración o frase en unidades más pequeñas, conocidas como tokens. Estos tokens pueden incluir palabras, fechas, signos de puntuación o incluso fragmentos de palabras.

## ¿Qué es la tokenización en NLP?

El procesamiento del lenguaje natural (NLP) es un subcampo de la informática, la inteligencia artificial, la ingeniería de la información y la interacción entre humanos y computadoras. Se centra en la programación de computadoras para procesar y analizar grandes cantidades de datos en lenguaje natural. La complejidad de leer y comprender idiomas hace que el PNL sea un campo desafiante.

La tokenización es un primer paso crucial en el proceso de PNL que influye en todo el flujo de trabajo.

La tokenización es el proceso de dividir el texto en unidades más pequeñas, o tokens. Estos tokens suelen ser palabras o subpalabras en el procesamiento del lenguaje natural. Es un paso fundamental en muchas tareas de PNL, incluido el procesamiento de texto, el modelado del lenguaje y la traducción automática. Los tokens sirven como bloques de construcción para un mayor procesamiento y análisis, lo que permite la conversión de texto no estructurado en una forma estructurada adecuada para el aprendizaje automático.

## Tipos de tokenización

La tokenización se puede clasificar en varios tipos según el método de segmentación de texto:

### Tokenización de palabras

- **Descripción**: Divide el texto en palabras individuales.
- **Ejemplo**:
Entrada: "La tokenización es una tarea importante de PNL".
Salida: ["Tokenización", "es", "una", "importante", "PNL", "tarea", "."]

### Tokenización de oraciones

- **Descripción**: Segmenta el texto en oraciones.
- **Ejemplo**:
Entrada: "La tokenización es una tarea importante de PNL. Ayuda a dividir el texto en unidades más pequeñas".
Salida: ["La tokenización es una tarea importante de NLP.", "Ayuda a descomponer el texto en unidades más pequeñas."]

### Tokenización de subpalabras

- **Descripción**: Descompone las palabras en unidades más pequeñas.
- **Ejemplo**:
Entrada: "tokenización"
Salida: ["token", "ización"]

### Tokenización de caracteres

- **Descripción**: Divide el texto en caracteres individuales.
- **Ejemplo**:
Entrada: "Tokenización"
Salida: ["T", "o", "k", "e", "n", "i", "z", "a", "t", "i", "o", "n"]

## Necesidad de tokenización

La tokenización juega un papel crucial en NLP por varias razones:

- **Procesamiento de texto eficaz**: Simplifica el texto sin formato para facilitar su procesamiento y análisis.

- **Extracción de características**: Permite la representación numérica de datos de texto para modelos de aprendizaje automático.

- **Modelado de lenguaje**: facilita representaciones organizadas del lenguaje para la generación de texto y el modelado de lenguaje.

- **Recuperación de información**: esencial para la indexación y búsqueda eficientes basadas en palabras o frases.

- **Análisis de texto**: admite tareas de procesamiento del lenguaje natural como el análisis de sentimientos y el reconocimiento de entidades con nombre.

- **Gestión de vocabulario**: ayuda a gestionar el vocabulario de un corpus generando una lista de tokens distintos.

- **Adaptación específica de la tarea**: se puede personalizar para adaptarse a tareas específicas de procesamiento del lenguaje natural, como el resumen y la traducción automática.

- **Paso de preprocesamiento**: transforma el texto sin formato en un formato adecuado para un análisis estadístico y computacional posterior.

## Expresiones regulares: Regex

Las expresiones regulares (regexes) son patrones que describen una cierta cantidad de texto. Por lo general, este patrón lo utilizan los algoritmos de búsqueda de cadenas para operaciones de "búsqueda" o "buscar y reemplazar" en cadenas, o para la validación de entrada.

### Metacharacters:

| Elemento | Descripción | Ejemplo de Regex | Texto de Ejemplo | Coincidencias |
|----------|-------------|------------------|------------------|---------------|
| `^`      | Inicio de línea | `^Hola` | Hola, mundo! | Hola |
| `$`      | Fin de línea | `mundo!$` | Hola, mundo! | mundo! |
| `.`      | Cualquier carácter (excepto nueva línea) | `H.la` | Hola, mundo! | Hola |
| `*`      | Cero o más del carácter anterior | `Ho*la` | Hoooooola, mundo! | Hoooooola |
| `+`      | Uno o más del carácter anterior | `Ho+la` | Hoooooola, mundo! | Hoooooola |
| `?`      | Cero o uno del carácter anterior | `H?ola` | Hola, mundo! | Hola |
| `{n}`    | Exactamente n veces del carácter anterior | `o{2}` | Hola, mundo! | oo (en Hoooooola) |
| `{n,}`   | n o más veces del carácter anterior | `o{2,}` | Hoooooola, mundo! | oooooo (en Hoooooola) |
| `{n,m}`  | Entre n y m veces del carácter anterior | `o{1,3}` | Hoooooola, mundo! | ooo (en Hoooooola) |
| `[abc]`  | Cualquiera de los caracteres entre corchetes | `[Hh]ola` | Hola, hola | Hola, hola |
| `[^abc]` | Cualquier carácter que NO esté entre corchetes | `[^Hh]ola` | Bola, Hola | Bola |
| `(abc)`  | Grupo de captura | `(Hola)` | Hola, Hola mundo! | Hola (dos veces) |
| `\|`     | Alternancia (or) | `Hola\|Mundo` | Hola, Mundo | Hola, Mundo |
| `\d`     | Cualquier dígito (equivalente a `[0-9]`) | `\d+` | La casa 123 | 123 |
| `\w`     | Cualquier carácter de palabra (letras, dígitos, _) | `\w+` | Hola_123 | Hola_123 |
| `\s`     | Cualquier espacio en blanco (espacio, tabulación, nueva línea) | `Hola\sMundo` | Hola Mundo | Hola Mundo |
| `\b`     | Límite de palabra | `\bHola\b` | Hola, mundo! | Hola |

## Stop Words

Las **stop words** son **palabras muy frecuentes** en un idioma que, por sí solas, **aportan poco o ningún significado relevante** para el análisis de texto.

En procesamiento de lenguaje natural (NLP), normalmente se **eliminan durante el preprocesamiento** para reducir ruido y mejorar el desempeño de los modelos.

---

### Definición
> **Stop words** son palabras funcionales que aparecen con alta frecuencia y cuyo valor semántico es bajo para tareas como clasificación, búsqueda o análisis de frecuencia.

---

### Ejemplos comunes

**En inglés**
- the, is, and, to, of, in, on, with, that

**En español**
- el, la, los, las, de, que, y, a, en, por

---

### ¿Por qué se eliminan?
- Reducen la dimensionalidad del vocabulario  
- Evitan que palabras poco informativas dominen el análisis  
- Mejoran modelos como **Bag of Words** y **TF-IDF**

---

### Importante: no siempre deben eliminarse
En algunas tareas, las stop words **sí son relevantes**, por ejemplo:
- Análisis de sentimientos (*not*, *no*, *never*)  
- Modelos de lenguaje (BERT, GPT)  
- POS tagging, NER y análisis sintáctico  

| Característica | Descripción |
|---------------|-------------|
| Frecuencia | Muy alta |
| Carga semántica | Baja |
| Uso en NLP | Generalmente se eliminan |
| Excepciones | Sentimiento, contexto, transformers |

```python
def quita_stopwords( vtokens, diccionario = "english"):
    palabras = stopwords.words(diccionario)
    return [token for token in vtokens if token not in palabras]
```

# Lematización en Procesamiento de Lenguaje Natural (NLP)

La **lematización** es una técnica fundamental en el procesamiento de lenguaje natural (NLP) que nos permite reducir las palabras a su **forma base o lema**, que es la forma estándar o de diccionario de la palabra. A diferencia del **stemming**, que recorta las terminaciones de las palabras sin considerar su contexto, la lematización toma en cuenta la **categoría gramatical** y el **contexto semántico**.

## ¿Por qué es importante la lematización?

En muchas tareas de NLP, como el análisis de sentimientos, la clasificación de textos o la búsqueda de información, una misma palabra puede aparecer en diversas formas. Por ejemplo:

- "corriendo", "corrió", "correrán", y "corre" son formas del verbo **"correr"**.
  
La lematización nos permite **reducir todas esas formas a una única palabra base ("correr")**, mejorando así la precisión y eficiencia en el análisis.

### Diferencia entre Lematización y Stemming

| Técnica        | Resultado para "corriendo" | Resultado para "hombres" |
|----------------|----------------------------|--------------------------|
| **Lematización** | correr                     | hombre                   |
| **Stemming**    | corr                        | hombr                    |

La lematización genera palabras con significado en el idioma, mientras que el stemming a veces produce palabras truncadas que no siempre son palabras válidas.

|-------------------------|------------------|----------------------------------|-----------------------------|

| TF-IDF (Term Frequency–Inverse Document Frequency) | Sí | Sí | Destaca palabras relevantes y evita duplicados de raíz. |

## Graficación de frecuencias

## WordCloud

Un **WordCloud** (nube de palabras) es una **visualización gráfica** que muestra las palabras más frecuentes de un texto o conjunto de documentos, donde el **tamaño de cada palabra es proporcional a su frecuencia** o peso.

Se utiliza principalmente como una técnica de **análisis exploratorio** para identificar rápidamente los términos más relevantes dentro de un corpus.

---

### ¿Qué información proporciona?
- Permite identificar **palabras dominantes** en el texto  
- Ayuda a detectar **temas recurrentes**  
- Facilita la interpretación inicial de grandes volúmenes de texto  

---

### ¿Cuándo es útil?
- Exploración inicial de datos textuales  
- Análisis de reseñas, opiniones o comentarios  
- Comparación visual entre distintos subconjuntos de texto  

---

### Limitaciones
- No muestra relaciones semánticas entre palabras  
- No conserva el orden del texto  
- No es adecuado para análisis cuantitativo profundo  

---

### Relación con NLP
Un WordCloud suele generarse **después del preprocesamiento**, utilizando texto tokenizado y limpio, y suele complementarse con técnicas como:
- Conteo de palabras  
- TF-IDF  
- Clasificación de textos  


```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
def nube_palabras(corpus, stopwords, color):
    stopwords = set(stopwords)
    wordcloud = WordCloud(
        background_color=color,
        stopwords=stopwords,
        max_words=100,
        max_font_size=50,
        random_state=1,
        collocations=False)

    wordcloud=wordcloud.generate(str(corpus))

    fig = plt.figure(1, figsize=(15, 15))
    plt.imshow(wordcloud)
    plt.axis('on')
    plt.show()
```
